## Transformation

### 1.Calendar

```python
df_cal=df_cal.withColumn('Month',month(col('Date'))).withColumn('Year',year(col('Date')))
df_cal.display()
```

```python
df_cal.write.format('parquet').mode('append').option("path","abfss://silver@awstorageaccount007.dfs.core.windows.net/AdventureWorks_Calendar").save()
```

###2.Customers

```python
df_cus=df_cus.withColumn("full_Name",concat_ws(" ",col("Prefix"),col("FirstName"),col("LastName")))
```

```python
df_cus.write.format('parquet').mode('append').option("path","abfss://silver@awstorageaccount007.dfs.core.windows.net/AdventureWorks_Customers").save()
```

###3.Product SubCategories

```python
df_prosubcat.write.format('parquet').mode('append').option("path","abfss://silver@awstorageaccount007.dfs.core.windows.net/AdventureWorks_Product_SubCategories").save()
```

###4.Products

```python
df_pros=df_pros.withColumn("ProductSKU",split(col("ProductSKU"),"-")[0]).withColumn("ProductName",split(col("ProductName")," ")[0])
```

```python
df_pros.write.format('parquet').mode('append').option("path","abfss://silver@awstorageaccount007.dfs.core.windows.net/AdventureWorks_Products").save()
```

###5.Returns

```python
df_ret.write.format('parquet').mode('append').option("path","abfss://silver@awstorageaccount007.dfs.core.windows.net/AdventureWorks_Returns").save()
```

###6.Territories

```python
df_ter.write.format('parquet').mode('append').option("path","abfss://silver@awstorageaccount007.dfs.core.windows.net/AdventureWorks_Territories").save()
```

###7.Sales

```python

df_sales=df_sales.withColumn("StockDate",to_timestamp('StockDate'))
```

```python
df_sales=df_sales.withColumn("OrderNumber",regexp_replace(col("OrderNumber"),"S","T"))
```

```python
df_sales=df_sales.withColumn("multiply",col("OrderLineItem")*col("OrderQuantity"))
```

```python
df_sales.write.format('parquet').mode('append').option("path","abfss://silver@awstorageaccount007.dfs.core.windows.net/AdventureWorks_Sales").save()
```

###8.Product Categories

```python
df_procat.write.format('parquet').mode('append').option("path","abfss://silver@awstorageaccount007.dfs.core.windows.net/AdventureWorks_Product_Categories").save()
```
